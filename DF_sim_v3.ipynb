```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import astropy.units
from astropy import constants as const
import collections, numpy
import matplotlib.lines as mlines
from matplotlib.ticker import AutoMinorLocator
```

```python
#Defining Variables and constants

Big_M = (1e8 * const.M_sun.value) #kg
Obj_Radius = 0.1 * const.au.value # m
vi = 2e9 #m/s
xi = 4*const.au.value
ymax = const.au.value  #m
#ypos=np.linspace(0,1,int(N)+1)[1:]*(ymax)


     
dt = 6 # second time step
N = 1000 #Number of steps 
 

colors = mpl.cm.get_cmap()(np.linspace(0.0, 1.0, int(N))) # setting colors for data pts


escape_vel= np.sqrt(2*const.G.value * Big_M/xi)
print(f"escape velocity = {escape_vel/1e5:.4e}")
escape_time = xi/(escape_vel * vi)
#print(f"escape time = {time:.4e}")
dt,ymax
```

```python

#setting y values for each small mass object ???
yvals = np.linspace(0.0, 1.0, int(N)+1)[1:] * ymax

#setting the same value of x for each small mass to start at in cm
xvals = np.ones_like(yvals) * xi

#Setting the velocity for each particle in the x direction. 
# Does this mean each particle starts out at the same velocity???
xvel = vi * np.ones_like(yvals)

#velocity in y direction. Since this is a 1-D problem I am assuming 
#that's why this velocity should be zero ???
yvel = np.zeros_like(yvals)
```

```python
time = np.array([0])      
xpos = np.array([xi])    
velocity = np.array([vi])
```

```python
t = 0
P=10
vy = np.zeros(P)
vx = np.ones(P)*(-vi)
y=np.linspace(0,1,int(P)+1)[1:]*(ymax)
#x = np.zeros(P)
x = np.ones(P)*xi

Time = np.zeros((P,N))     
Xpos = np.zeros((P,N))
Ypos = np.zeros((P,N))
Yvel = np.zeros((P,N))
Xvel = np.zeros((P,N))
print('          x', '               y')
for i in range(N): #Avoid square arrays!!!
    print("i=", i)
    for j in range(P):
        print("j=",j)
        # here are the calculations
        t = t + dt  #incrementing time value by dt
        A_x = ((-const.G.value*Big_M/((x[j]**2+y[j]**2)**(3/2)))*x[j])
        A_y = ((-const.G.value*Big_M/((x[j]**2+y[j]**2)**(3/2)))*y[j])
        #print(A_x,A_y)
        #continue
        vx[j] = vx[j] + A_x * dt 
        vy[j] = vy[j] + A_y * dt
        #print((vx),(vy))
        #continue
        x[j] = x[j] + vx[j] * dt
        y[j] = y[j] + vy[j] * dt 
        #print(x)
        #continue
        Time[j,i] = t
        Xpos[j,i] = x[j]
        Ypos[j,i] = y[j]
        Xvel[j,i] = vx[j]
        Yvel[j,i] = vy[j]
        #print((Xvel))
        #continue
    print(Xpos[:,i])
    #break
```

```python
#Defining Variables and constants

Obj_M = (1e8 * const.M_sun.value) #kg
Obj_R = 0.1 * const.au.value # m
init_vel = 2e9 #m/s
x_pos = 4*const.au.value
y_max = const.au.value  #m
y_pos=np.linspace(0,1,int(P)+1)[1:]*(y_max)


     
DT = 1e1 # second time step
m = 1000 #Number of steps 
Particles = 10 

colors = mpl.cm.get_cmap()(np.linspace(0.0, 1.0, int(N))) # setting colors for data pts


esc_velocity= np.sqrt(2*const.G.value * Obj_M/init_vel)
print(f"escape velocity = {esc_velocity/1e5:.4e}")
esc_time = init_vel/(esc_velocity * vi)
#print(f"escape time = {time:.4e}")
DT,y_max
```

```python
#Defining equation for Potential energy of each particle for N=1000
U = np.zeros((Particles,m))
K = np.zeros((Particles,m))
def UE(Xpos,Ypos):
    Pot=(-const.G.value*Obj_M)/(np.sqrt(Xpos**2+Ypos**2))
    return Pot
UE(Xpos, Ypos)
```

```python
np.set_printoptions(threshold=np.inf)
print((UE(Xpos,Ypos)))
```

```python
#Defining equation for Kinetic energy of each particle for N=1000
def TE(Xvel, Yvel):
    T=(0.5*(Xvel+Yvel)**2)
    return T
TE(Xvel,Yvel)
```

```python
#Defining equation for the Total energy of each particle for N=1000
#We use the specific kinetic and potential energy equations 
def TotalE(Xpos,Ypos,Xvel,Yvel):
    TotE=(-const.G.value*Obj_M)/(np.sqrt(Xpos**2+Ypos**2))+(0.5*(Xvel**2+Yvel**2))
    return TotE

TotalE(Xpos,Ypos,Xvel,Yvel)
```

```python
#Percentage of energy conserved 1= all energy conserved. E/E_0
Percentage_E=np.zeros((P,N))
for i in range(N):
    for j in range(P):
        TotE_P=TotalE(Xpos,Ypos,Xvel,Yvel)[j,i]/TotalE(Xpos,Ypos,Xvel,Yvel)[j,0]
        Percentage_E[j,i]=TotE_P
(Percentage_E)
```
